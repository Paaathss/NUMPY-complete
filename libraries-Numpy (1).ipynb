COPY AND VIEW

```python
import numpy as np

x = np.array([2,45,8,81,5,3])

y = x

z = x.copy()   #deep copy

w = x.view()   #shallow copy

print(w)
print(x)
print(y)
print(z)

x[2] = 100

print(w)
print(x)
print(y)
print(z)
```

ARRAY JOINING


```python
#vertical stack

import numpy as np

a = np.array([[3,4,2],[4,5,3]])

b= np.array([[4,3,1],[5,32,0]])

c = np.vstack((a,b))

print(c)
```

```python
#horizontal stack

import numpy as np

a = np.array([[3,4,2],[4,5,3]])

b= np.array([[4,3,1],[5,32,0]])

c = np.hstack((a,b))

print(c)
```

```python
#column stack

import numpy as np

a = np.array([[3,4,2],[4,5,3]])

b= np.array([[4,3,1],[5,32,0]])

c = np.column_stack((a,b))

print(c)
```

```python
#row stack

import numpy as np

a = np.array([[3,4,2],[4,5,3]])

b= np.array([[4,3,1],[5,32,0]])

c = np.row_stack((a,b))

print(c)
```

CONCATENATE



```python
import numpy as np

a = np.array([[3,4,2],[4,5,3]])

b= np.array([[4,3,1],[5,32,0]])

c = np.concatenate((a,b))

print(c)

```
